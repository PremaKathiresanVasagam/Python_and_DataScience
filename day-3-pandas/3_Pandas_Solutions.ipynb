Import the first dataset cars1 and cars2 and assign it to variable cars1 and cars2.

https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv


```python
cars1 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars1.csv")
cars2 = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/05_Merge/Auto_MPG/cars2.csv")

print(cars1.head())
print(cars2.head())
```

Oops, it seems our first dataset has some unnamed blank columns, fix cars1 by removing nan columns

```python
cars1 = cars1.loc[:, "mpg":"car"]
cars1.head()
```

Join cars1 and cars2 into a single DataFrame called cars.

Clue: try pd.append()

```python
cars = cars1.append(cars2)
cars
```

Create a random number Series from 15,000 to 73,000 in car_owners variable.

```python
car_owners = np.random.randint(15000, high=73001, size=398, dtype='l')
car_owners
```

Add the column owners to cars
